```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
textblob = pd.read_csv('sentiment/data_textblob.csv')
vader = pd.read_csv('sentiment/data_vader.csv')
```

```python
def plot_sentiment_bar(sentiment_values, name, sentimant):
    # Set the bins for the histogram
    bins = np.linspace(-1, 1, num=11)  # Divide the range [-1, 1] into 10 equal bins

    # Calculate the histogram of sentiment values
    counts, bin_edges = np.histogram(sentiment_values, bins=bins)

    # Create the x-axis values
    x_values = [(bin_edges[i] + bin_edges[i+1]) / 2 for i in range(len(bin_edges)-1)]

    # Plot the bar graph
    plt.bar(x_values, counts, width=0.2)

    # Set labels and title
    plt.xlabel('Sentiment Value')
    plt.ylabel('Count')
    plt.title(f"Sentiment Analysis for {name} with {sentimant}")

    target_directory = f'./bar graph/{sentimant}'
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)
    plt.savefig(os.path.join(target_directory, f"{name}.png"), transparent=True)

    plt.show()
```

```python
for col in vader.columns[5:]:
    plot_sentiment_bar(vader[col], name=col, sentimant="vader")
```

```python
for col in textblob.columns[5:]:
    plot_sentiment_bar(textblob[col], name=col, sentimant="textblob")
```
